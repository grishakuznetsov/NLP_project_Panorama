# Групповой проект

# 1) Сбор датасета

Выкачаем n новостей со страницы газеты "Панорама", в датасет сохраним 

Импортируем нужные для работы модули

```
import pandas as pd
import numpy as np
import requests
import re
from tqdm.auto import tqdm

from fake_useragent import UserAgent
ua = UserAgent(verify_ssl=False)

headers = {'User-Agent': ua.random}
```

```
pip install fake_useragent
```

```
headers_panorama = []
articles_panorama = []
links_panorama = []
tags_panorama = []
newspapers_p = []
topics, topics_bd = ['society', 'economics', 'science', 'politics'], ['Общество', 'Экономика', 'Наука', 'Политика']
```

```
for t in range(len(topics)):
    for i in tqdm(range(10)):
        if i == 1:
            page = "https://panorama.pub/category/news/{0}".format(topics[t])
        else:
            page = "https://panorama.pub/category/news/{0}/page/{1}".format(topics[t], i)
        result = requests.get(page, headers=headers)
        html = result.text
        l = re.findall(r'bookmark">(.*?)</a></h2>', str(html))
        inf = re.findall(r'<h2 class="entry-title"><a href="(.*?)" rel="bookmark">(.*?)</a></h2>', html)
        for i in tqdm(range(len(inf))):
            result_link = requests.get(inf[i][0], headers=headers)
            html_link = result_link.text
            cur_text = re.findall(r'</div><p>(.*)</p>.<div', html_link, flags = re.DOTALL)
            p = re.split('</p>\n<p>', ' '.join(cur_text))
            clean_text = re.sub(r'[a-z;&0-9<>\|/=?!#&"_-]*', '', ' '.join(p))
            article_ok = re.sub(r'(:?(:..)| ,|\t)', '', ''.join(clean_text))
            article_to_list = '. '.join(re.split(r'\. ', article_ok))
            if len(article_to_list) >= 100:
              links_panorama.append(inf[i][0])
              headers_panorama.append(''.join(re.sub('[a-z;&0-9<>\|/=#&_-]*','',inf[i][1])))
              articles_panorama.append(article_to_list)
              tags_panorama.append(topics_bd[t])
              newspapers_p.append('Panorama')

```

```
data_of_news_p = pd.DataFrame({'newspaper': newspapers_p,
                'tags': tags_panorama,
                'headers': headers_panorama,
                'articles': articles_panorama,
                'links': links_panorama})
```

```
data_of_news_p.to_csv('data_of_news_p.csv')
from google.colab import files
files.download('data_of_news_p.csv') 
```
